```python
import tkinter as tk
from tkinter import ttk

LARGE_FONT= ("Verdana", 12)

class Scraper(tk.Tk):

    def __init__(self, *args, **kwargs):
        
        tk.Tk.__init__(self, *args, **kwargs)
        tk.Tk.wm_title(self, "Tasky !")
        
        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand = True)
        container.grid_rowconfigure(0, weight=1)
        container.grid_columnconfigure(0, weight=1)

        

        menubar = tk.Menu(container)
        filemenu = tk.Menu(menubar, tearoff=0)
        filemenu.add_command(label="Save settings", command=lambda: popupmsg('Not supported just yet!'))
        filemenu.add_separator()
        filemenu.add_command(label="Exit", command=quit)
        menubar.add_cascade(label="File", menu=filemenu)


        pomodora = tk.Menu(menubar, tearoff=1)
        pomodora.add_command ( label="Add task",
                                  command=lambda: changeExchange('BTC-e','btce') )
        pomodora.add_command ( label="List Of Previous Tasks",
                                  command=lambda: changeExchange('Bitfinex','bitfinex') )
        pomodora.add_separator()
        pomodora.add_command ( label="Settings",
                                  command=lambda: changeExchange('Huobi','huobi') )
        menubar.add_cascade(label = "Pomodora", menu = pomodora)



        web_blocker = tk.Menu(menubar, tearoff=1)
        web_blocker.add_command ( label="Add Websites",
                                  command=lambda: changeTimeFrame('tick') )
        web_blocker.add_command ( label="List Websites",
                                  command=lambda: changeTimeFrame('1d') )
        web_blocker.add_separator()
        web_blocker.add_command ( label="Settings",
                                  command=lambda: changeTimeFrame('7d') )
        menubar.add_cascade(label = "Website Blocker", menu = web_blocker)


        bot_manager = tk.Menu(menubar, tearoff=1)

        bot_manager.add_command ( label="My Bots",
                                  command=lambda: changeTimeFrame('tick') )
        bot_manager.add_command ( label="Make New Bots !",
                                  command=lambda: changeSampleSize('1Min',0.0005) )
        bot_manager.add_separator()
        bot_manager.add_command ( label="Settings",
                                  command=lambda: changeSampleSize('3H',0.096) )
        menubar.add_cascade(label = "Botsmanager", menu = bot_manager)


        manager = tk.Menu(menubar, tearoff=1)
        manager.add_command ( label="My Managers",
                                  command=lambda: print('NOT live yet'))
        manager.add_command ( label="Add new Manager",
                                  command=lambda: print('NOT live yet'))

#         manager.add_separator()
#         manager.add_command ( label="Quick Buy",
#                                   command=lambda: print('quick buy!'))#, accelerator="Ctrl+B" )
#         manager.add_command ( label="Quick Sell",
#                                   command=lambda: print('quick sell'))#, accelerator="Ctrl+S")
        manager.add_separator()
        manager.add_command ( label="Settings",
                                  command=lambda: print('quick buy!'))
        menubar.add_cascade(label = "Manager", menu = manager)    
        

        helpmenu = tk.Menu(menubar, tearoff=0)
        helpmenu.add_command(label="Tutorial",) #command=tutorial)
        menubar.add_cascade(label="Help", menu=helpmenu)

        tk.Tk.config(self, menu=menubar)
        

        self.frames = {}
        for F in (StartPage, TaskPage):

            frame = F(container, self)
            self.frames[F] = frame
            frame.grid(row=0, column=0, sticky="nsew")

        self.show_frame(StartPage)

#         tk.Tk.iconbitmap(self,default='clienticon.ico')

    def show_frame(self, cont):

        frame = self.frames[cont]
        frame.tkraise()

        
class StartPage(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self,parent)
        label = tk.Label(self, text="Start Page", font=LARGE_FONT)
        label.pack(pady=10,padx=10)
        
class TaskPage(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        label = ttk.Label(self, text="BTC-e Exchange Page", font=LARGE_FONT)
        label.pack(pady=10,padx=10)



app = Scraper()
app.geometry("1280x720")
app.mainloop()
```
